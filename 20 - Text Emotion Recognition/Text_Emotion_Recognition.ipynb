### Import Libraries

```python
import transformers
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

### Use pre-trained model

```python
tokenizer = AutoTokenizer.from_pretrained("bergum/xtremedistil-l6-h384-go-emotion")

model = AutoModelForSequenceClassification.from_pretrained("bergum/xtremedistil-l6-h384-go-emotion")
```

### Predict

```python
pipeline = transformers.pipeline("text-classification",
                                 model=model,
                                 tokenizer=tokenizer)
```

```python
emotion = str(input("Enter a sentence: \n"))
result = pipeline(emotion)
print('\nEmotion: ', result[0]['label'], '\n')
print('Score: ', result[0]['score'])
```
