### Import Libraries

```python
from transformers import pipeline, set_seed
```

```python
generator_bloom = pipeline('text-generation',
                           model='bigscience/bloom-560m',
                           device=0,
                           use_auth_token=True)
```

```python
kwargs = {
    "max_length": 200,
    "do_sample": True,
    "top_k": 30,
    "top_p": 0.9,
    "temperature": 0.3,
    "repetition_penalty": 2.0    
}
```

### Prompt

```python
text = "Smart contracts"
seed = 10
```

```python
%%time

set_seed(seed)
result_bloom = generator_bloom(text, **kwargs)
print(f"{result_bloom[0]['generated_text']}\n")
```
